Cargamos todas las tablas de la página de Wikipedia sobre las Elecciones Presidenciales 2015

```python
import pandas as pd
url = "https://es.wikipedia.org/wiki/Elecciones_presidenciales_de_Argentina_de_2015"
res = pd.read_html(url)
```

Seleccionamos las tablas de encuestas para las elecciones generales:

```python
pre_paso = res[4]
post_paso = res[5]
```

A ver cómo quedaron...

```python
pre_paso
```

Todo en orden, salvo el encabezado:

```python
pre_paso.columns = pre_paso.loc[0]
pre_paso.drop(0,inplace=True)
pre_paso
```

 A ver la otra...

```python
post_paso
```

Hay que arreglar algunas filas al principio (y el encabezado)

```python
post_paso.columns = post_paso.loc[0]
post_paso.drop(0,inplace=True)

fila = post_paso.loc[2]
post_paso.loc[2] = ['08/2015'] + fila.tolist()[:-1]

post_paso.loc[3] = post_paso.loc[2].tolist()[:2] + post_paso.loc[3].tolist()[:-2]
post_paso.loc[4] = ['08/2015'] + post_paso.loc[4].tolist()[:-1]

post_paso
```
